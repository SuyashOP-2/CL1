```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("Wine.csv")
data.head()

scaler = StandardScaler()
wine_data_scaled = scaler.fit_transform(data)

pca = PCA(n_components = 2)
pca.fit(wine_data_scaled)

wine_data_pca = pca.transform(wine_data_scaled)
wine_data_pca

plt.scatter(wine_data_scaled[:,0], wine_data_scaled[:,1], c='b')
plt.colorbar(label="Wine Type")
plt.title("Wine data PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
